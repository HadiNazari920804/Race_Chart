```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
%matplotlib notebook
```

```python
df=pd.read_csv('DGB.csv',index_col='todate',encoding='utf8')
df
```

```python
branch_list=df.columns
todates=np.sort(df.index)
```

```python
i=11
fig,ax=plt.subplots(figsize=(10,10))
ax.barh(y=df.iloc[i].rank(method='first'),width=df.iloc[i].values,tick_label=df.iloc[i].index,
        align='center',color=plt.cm.tab10(range(len(branch_list))))
ax.set_title(df.index[i])
```

```python
#db.index = range(0,(len(years))*10,10)
#print(list(db.index))
```

```python
l1=[]
for i in range(0,(len(todates))*20,20):
    l1.append(19+i)
df.index=l1
```

```python
df
```

```python
row_nums = [i for i in range(0,len(todates)*20) if i not in l1 ]
empty = pd.DataFrame(np.nan, index= row_nums, columns = df.columns)
empty.head(11)
```

```python
expand_df = pd.concat([df, empty]).sort_index()
expand_df.iloc[0]=0
expand_df.head(11)
```

```python
rank_df = expand_df.rank(axis=1,method='first')
rank_df.head(11)
```

```python
expand_df = expand_df.interpolate()
expand_df.head(11)
```

```python
rank_df = rank_df.interpolate()
rank_df.head(11)
```

```python
l=[]
for d in todates:
    t=[d]*20
    l.append(t)
flat_list = []
for sublist in l:
    for item in sublist:
        flat_list.append(item)
```

```python
expand_df.index=flat_list
```

```python
expand_df
```

```python
def update(i):
    ax.clear()
    ax.set_facecolor(plt.cm.Greys(0.2))
    [spine.set_visible(False) for spine in ax.spines.values()]
    hbars = ax.barh(y = rank_df.iloc[i].values,
           tick_label=expand_df.iloc[i].index,
           width = expand_df.iloc[i].values,
           height = 0.8,
           color = plt.cm.tab10(range(11))
           )
    ax.set_title(flat_list[i])
    ax.bar_label(hbars, labels=["" if x==0 else f'{x:.2%}' for x in hbars.datavalues])
    ax.set_xticks([])

fig,ax = plt.subplots(#figsize=(10,7),
                      facecolor = plt.cm.Greys(0.2),
                      dpi = 150,
                      tight_layout=True
                     )


data_anime = FuncAnimation(
    fig = fig,
    func = update,
    frames= len(expand_df),
    interval=140, repeat=False
)

FFwriter = animation.FFMpegWriter(fps=7)
data_anime.save('animation.mp4', writer = FFwriter)

plt.show()
```
